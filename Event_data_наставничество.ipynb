```python
import pandas as pd
import numpy as np
```

```python
from datetime import date
```

```python
df=pd.read_csv('event_data.csv')
```

```python
df
```

```python
df['event_date']=pd.to_datetime(df['event_date'], format='%Y-%m-%d')
```

```python
df.dtypes
```

```python
visit_first_activity_date = df.groupby(['user_id'])['event_date'].min()
visit_first_activity_date.name = 'first_activity_date'
display(visit_first_activity_date.head())
```

```python
visit_first_activity_date_gr = df.groupby('user_id').agg({'event_date':'min'}).reset_index()
#visit_first_activity_date.name = 'first_activity_date'
visit_first_activity_date_gr = visit_first_activity_date_gr.set_index('user_id')
visit_first_activity_date_gr.head()
```

```python
df = df.join(visit_first_activity_date, on='user_id')
display(df.head())
```

```python
df['activity_week'] = df['event_date'].dt.week
df['first_activity_week'] = df['first_activity_date'].dt.week                                 
display(df.head())
```

```python
df.groupby(['first_activity_week'])['user_id'].agg(['count','nunique'])
```

```python
df['cohort_lifetime'] = df['event_date'] - df['first_activity_date']
```

```python
df
```

```python
df['cohort_lifetime'] = df['cohort_lifetime'] / np.timedelta64(1,'W')
df['cohort_lifetime'] = df['cohort_lifetime'].round().astype('int')
display(df.head())
```

```python
cohorts = df.groupby(['first_activity_week', 'cohort_lifetime']).agg({'user_id':'nunique'}).reset_index() 
display(cohorts.head())
display(cohorts.tail())
```

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_week','user_id']]
initial_users_count = initial_users_count.rename(columns={'user_id':'cohort_users'}) 
display(initial_users_count) 
```

```python
cohorts = cohorts.merge(initial_users_count,on='first_activity_week')
display(cohorts.head())
```

```python
cohorts['retention'] = cohorts['user_id']/cohorts['cohort_users'] 
display(cohorts.head())
```

```python
retention_pivot = cohorts.pivot_table(index='first_activity_week',columns='cohort_lifetime',values='retention',aggfunc='sum')
display(retention_pivot)
```

```python
ARPPU = df.groupby(['first_activity_week', 'cohort_lifetime']).agg({'user_id':'nunique','purchase_amount':'sum'}).reset_index() 
```

```python
ARPPU['ARPPU']=ARPPU['purchase_amount']/ARPPU['user_id']
```

```python
ARPPU_pivot = ARPPU.pivot_table(index='first_activity_week',columns='cohort_lifetime',values='ARPPU',aggfunc='sum')
```

```python
display(ARPPU_pivot)
```
